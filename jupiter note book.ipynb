```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,10,0.1)
x
```

```python
y = x*x
z =x+x
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.plot(x,z)
```

```python
plt.figure(figsize=(12,5))
plt.plot(x,y,c='red',label='x vs y')
plt.plot(x,z,c='green',label='x vs z')
plt.legend()
```

```python
plt.figure(figsize=(15,5))
plt.scatter(x,y)
plt.legend()
plt.grid(True)
plt.show()
```
